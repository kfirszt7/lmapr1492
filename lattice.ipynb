```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.core.lattice import Lattice
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

import numpy.matlib as np
import math
import pymatgen as mp
import pymatgen.symmetry.analyzer as span

with MPRester("r7QKrVN0WzAjMGwDF") as m:
    structure = m.get_structure_by_material_id("mp-15793")
print(structure)

# primitive cell et vecteur de base du réseau direct et volume
prim = structure.get_primitive_structure()
print(prim)
cell = Lattice.rhombohedral(6.923143,34.340415)
print("Coordonnée des vecteurs :",cell)
vect = cell.lengths
#a = [cell.a,0,0]
#b = [0,cell.b,0]
#c = [0,0,cell.c]

print("Vecteur du réseau direct :",vect)

# réseau réciproque  
#pi = math.pi
#a1 = ((2*pi)/vol)*(np.cross(b,c))
#b1 = ((2*pi)/vol)*(np.cross(c,a))
#c1 = ((2*pi)/vol)*(np.cross(a,b))
#prob a*a1 != 2*pi
recip = cell.reciprocal_lattice
print("Coordonnée des vecteurs du réseau réciproque :",recip)
vectrecip = recip.lengths

#volrecip = np.inner(a1,np.cross(b1,c1))

print("Vecteur du réseau réciproque :",vectrecip)
#print("Volume du réseau réciproque :",volrecip)

#Type de maille
struc = span.SpacegroupAnalyzer(structure)
print("Type de maille :", struc.get_lattice_type())
print(" ")
#Système cristallin
print("Système cristallin :",struc.get_crystal_system())
print(" ")
#Groupe ponctuel
print("Groupe ponctuel :", struc.get_space_group_symbol())
print(" ")


file = CifWriter(structure)
file.write_file('Lattice_LiTbSe2.cif')
```
